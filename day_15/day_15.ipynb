```python
def get_robot_pos(grid):
    for i in range(rows):
        for j in range(cols):
            if grid[i][j] == "@":
                return i, j
get_robot_pos(grid)
```

```python
def calcuate_gps(grid):
    sum_gps = 0
    for i in range(rows):
        for j in range(cols):
            if grid[i][j] == "O":
                sum_gps += (100*i) + j
    return sum_gps
```

```python
with open("test_input.txt") as f:
# with open("input.txt") as f:
    lines = f.read().split("\n\n")
    
replace_dict = {
    "#": "##",
    "O": "[]",
    ".": "..",
    "@": "@."
}
def replace_with_dict(text, replacements):
    for key, value in replacements.items():
        text = text.replace(key, value)
    return text

new_grid = [list(replace_with_dict(line, replace_dict)) for line in lines[0].split("\n")]
rows = len(new_grid)
cols = len(new_grid[0])
```
