```python
with open("input.txt") as f:
    lines = f.readlines()
lines
```

```python
graph = [list(line.strip()) for line in lines]
rows = len(graph)
cols = len(graph[0])
rows, cols
```

```python
# graph
```

```python
horizontal_counts = 0
for row_num in range(rows):
    row = graph[row_num]
#     print("".join(row))
    for i in range(cols - 3):
        window = "".join(row[i : i + 4])
#         print("\t", window)
        if window in ["XMAS", "SAMX"]:
            horizontal_counts += 1
horizontal_counts
```

```python
vertical_counts = 0
for col_num in range(cols):
    col = ""
    for row in range(rows):
        col += graph[row][col_num]
        
#     print(col)
    for i in range(rows - 3):
        window = "".join(col[i : i + 4])
#         print("\t", window)
        if window in ["XMAS", "SAMX"]:
            vertical_counts += 1
vertical_counts
```

```python
diagonal_counts = 0
for i in range(rows - 3):
    for j in range(cols - 3):
        window = ""
        window += graph[i][j]
        window += graph[i + 1][j + 1]
        window += graph[i + 2][j + 2]
        window += graph[i + 3][j + 3]

        window_reverse = ""
#         window_reverse += graph[rows - 1 - i][cols - 1 - j]
#         window_reverse += graph[rows - 2 - i][cols - 2 - j]
#         window_reverse += graph[rows - 3 - i][cols - 3 - j]
#         window_reverse += graph[rows - 4 - i][cols - 4 - j]

        window_reverse += graph[i][cols - 1 - j]
        window_reverse += graph[i + 1][cols - 2 - j]
        window_reverse += graph[i + 2][cols - 3 - j]
        window_reverse += graph[i + 3][cols - 4 - j]
        
#         print(i, j, window)
#         print(i, j, window_reverse)
        
        if window in ["XMAS", "SAMX"]:
            diagonal_counts += 1
            
        if window_reverse in ["XMAS", "SAMX"]:
            diagonal_counts += 1
            
#     print()
diagonal_counts
```

```python
print(horizontal_counts + vertical_counts + diagonal_counts)
```
