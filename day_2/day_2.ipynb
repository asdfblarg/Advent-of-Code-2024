```python
import pandas as pd
```

```python
input_df = pd.read_csv("input.txt", names=["levels"])
input_df
```

```python
input_df["levels"] = input_df["levels"].str.split(" ").apply(lambda x: [int(num) for num in x])
input_df
```

```python
# input_df["levels_sorted"] = input_df["levels"].apply(sorted)
# input_df
```

```python
def check_all_increasing_or_decreasing(level):
    all_increasing = True
    all_decreasing = True
    for i in range(1, len(level)):
        if level[i-1] > level[i]:
            all_increasing = False
            
        if level[i-1] < level[i]:
            all_decreasing = False
    return all_increasing or all_decreasing

print(check_all_increasing_or_decreasing([1,2,3,4]))
print(check_all_increasing_or_decreasing([1,2,5,4]))
print(check_all_increasing_or_decreasing([4,5,2,1]))
print(check_all_increasing_or_decreasing([4,3,2,1]))
```

```python
def check_diff(level):
    for i in range(1, len(level)):
        if not 1 <= abs(level[i-1] - level[i]) <= 3:
            return False
    return True

print(check_diff([1,2,3,4]))
print(check_diff([1,2,6,4]))
print(check_diff([4,6,2,1]))
```

```python
input_df["all_incr_decre"] = input_df["levels"].apply(check_all_increasing_or_decreasing)
input_df["diff_ok"] = input_df["levels"].apply(check_diff)
input_df
```

```python
input_df["level_safe"] = input_df["all_incr_decre"] & input_df["diff_ok"]
input_df
```

```python
input_df[input_df["level_safe"]]
```

# part 2

```python
def check_diff_damper(level):
    for i in range(len(level)):
        new_level = level.copy()
        new_level.pop(i)
        all_incr_decre = check_all_increasing_or_decreasing(new_level)
        diff_ok = check_diff(new_level)
        if all_incr_decre and diff_ok:
            return True
    return False

print(check_diff_damper([1,2,3,4]))
print(check_diff_damper([1,3,2 4 5]))
print(check_diff_damper([9,7,6,2,1]))
```

```python
input_df["level_damper_safe"] = input_df["levels"].apply(check_diff_damper)
input_df
```

```python
input_df[input_df["level_damper_safe"]]
```
