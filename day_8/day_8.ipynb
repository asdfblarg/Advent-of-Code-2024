```python
from collections import defaultdict
from itertools import combinations
```

```python
# with open("test_input.txt") as f:
# with open("test_input_2.txt") as f:
with open("input.txt") as f:
    lines = f.readlines()
lines = [line.strip() for line in lines]
# lines
```

```python
grid = [list(line) for line in lines]
rows = len(grid)
cols = len(grid[0])
rows, cols
```

```python
def print_grid(grid):
    for row in grid:
        print("".join(row))


print_grid(grid)
```

```python
freqs = defaultdict(list)
freqs_list = []
for i in range(rows):
    for j in range(cols):
        if grid[i][j] != ".":
            freqs[grid[i][j]].append((i, j))
            freqs_list.append((i, j))
# freqs
```

```python
for freq in freqs:
    print(freq, freqs[freq])
```

```python
def generate_antinodes(freq_list):
    antinodes = set()
    for combo in combinations(freq_list, 2):
        a, b = combo

        x_diff = b[0] - a[0]
        y_diff = b[1] - a[1]

        antinodes.add((b[0] + x_diff, b[1] + y_diff))
        antinodes.add((a[0] - x_diff, a[1] - y_diff))

    return antinodes


# print(generate_antinodes(freqs["A"]))
print(sorted(generate_antinodes(freqs["0"])))
```

```python
antinodes = []
for freq in freqs:
    #     print(freq, freqs[freq])
    antinodes += generate_antinodes(freqs[freq])

len(set(antinodes))
```

```python
antinode_grid = [list(line) for line in lines]
valid_antinodes = set()
for antinode in antinodes:
    if (
        0 <= antinode[0] < rows and 0 <= antinode[1] < cols
    ):  # and antinode not in freqs_list:
        antinode_grid[antinode[0]][antinode[1]] = "#"
        valid_antinodes.add(antinode)

print_grid(antinode_grid)
```

```python
def generate_antinodes_2(freq_list):
    antinodes = set()
    for combo in combinations(freq_list, 2):
        a, b = combo

        x_diff = b[0] - a[0]
        y_diff = b[1] - a[1]

        cur = a
        while 0 <= cur[0] < rows and 0 <= cur[1] < cols:
            antinodes.add((cur[0] - x_diff, cur[1] - y_diff))
            cur = (cur[0] - x_diff, cur[1] - y_diff)

        cur = b
        while 0 <= cur[0] < rows and 0 <= cur[1] < cols:
            antinodes.add((cur[0] + x_diff, cur[1] + y_diff))
            cur = (cur[0] + x_diff, cur[1] + y_diff)

    return antinodes


# print(generate_antinodes(freqs["A"]))
# print(sorted(generate_antinodes_2(freqs["T"])))
```
