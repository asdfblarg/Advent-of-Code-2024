```python
from collections import defaultdict
```

```python
# with open("test_input.txt") as f:
with open("input.txt") as f:
    lines = f.readlines()
# lines
```

```python
lines = [line.strip() for line in lines]
# lines
```

```python
for i in range(len(lines)):
    if lines[i] == "":
        break_index = i
break_index
```

```python
rules = lines[:break_index]
rules
```

```python
orders = lines[break_index + 1 :]
orders = [order.split(",") for order in orders]
orders
```

```python
rules_dict = defaultdict(list)
for rule in rules:
    x, y = rule.split("|")
    rules_dict[x].append(y)
# rules_dict
```

```python
def check(order):
    # print(order)

    prev_pages = set()
    for i, page_num in enumerate(order):
        #     print(page_num)
        for bad_page in rules_dict[page_num]:
            if bad_page in prev_pages:
                # print("bad", bad_page, prev_pages, rules_dict[page_num])
                return False
        prev_pages.add(page_num)
    return True


check(orders[3])
```

```python
sum_middle = 0
for order in orders:
    # print(order, check(order))
    if check(order):
        middle_page = len(order) // 2
        # print(len(order), middle_page, int(order[middle_page]))
        sum_middle += int(order[middle_page])
sum_middle
```

```python
bad_orders = []
for order in orders:
    if not check(order):
        bad_orders.append(order)
bad_orders
```

```python
# order = bad_orders[2]
order = ["97", "75", "47", "29", "13"]
print(order)
prev_pages = set()
for i, page_num in enumerate(order):
    for bad_page in rules_dict[page_num]:
        if bad_page in prev_pages:
            print(page_num, "bad", bad_page, prev_pages, rules_dict[page_num])

    prev_pages.add(page_num)
```

```python
def get_bad_pages(order):
    prev_pages = {}
    for i, page_num in enumerate(order):
        for bad_page in rules_dict[page_num]:
            if bad_page in prev_pages:
                #                 print(page_num, "bad", bad_page, prev_pages, rules_dict[page_num])
                return (prev_pages[bad_page], i)

        prev_pages[page_num] = i


get_bad_pages(["97", "75", "47", "29", "13"])
```

```python
order = bad_orders[2]
print(order)
while not check(order):
    x, y = get_bad_pages(order)
    order[x], order[y] = order[y], order[x]
print(order)
print(check(order))
```

```python
corrected_bad_orders = []
for order in bad_orders:
    while not check(order):
        x, y = get_bad_pages(order)
        order[x], order[y] = order[y], order[x]
    corrected_bad_orders.append(order)

# corrected_bad_orders
```
