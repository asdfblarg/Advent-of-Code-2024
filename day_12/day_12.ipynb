```python
grid = [list(line) for line in lines]

def area(region, r, c):
    if (
        not (
            0 <= r < len(grid)
            and 0 <= c < len(grid[0])
            and (r, c) not in seen
            and grid[r][c]
        )
    ):
        return 0
    if grid[r][c] != region:
        return 0
    seen.add((r, c))
    regions.append((r, c))
    return (
        1
        + area(region, r + 1, c)
        + area(region, r - 1, c)
        + area(region, r, c - 1)
        + area(region, r, c + 1)
    )


seen = set()
regions = []

region_areas = []
for r in range(len(grid)):
    for c in range(len(grid[0])):
        if grid[r][c] == -1:
            continue
        else:
            region_start = grid[r][c]
            area_val = area(region_start, r, c)
            for cell in seen:
                grid[cell[0]][cell[1]] = -1
            region_areas.append((area_val, region_start, r, c))    

region_areas
```

```python
grid = [list(line) for line in lines]
def islandPerimeter(region, x, y):
    queue = deque()
    queue.append(tuple([x, y]))

    directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
    seen = set()
    total_perimeter = 0
    while queue:
        x, y = queue.popleft()
        if (x, y) in seen:
            continue
        if grid[x][y] == region:
            perimeter = 4
            seen.add((x, y))
        for direction in directions:
            new_x, new_y = x + direction[0], y + direction[1]
            if 0 <= new_x < rows and 0 <= new_y < cols:
                new_plot = grid[new_x][new_y]
                if new_plot == region:
                    queue.append((new_x, new_y))
                    perimeter -= 1
        total_perimeter += perimeter
                    
    for x,y in seen:
        grid[x][y] = -1
        
    return total_perimeter

# islandPerimeter("R", 0, 0)
```

```python
grid = [list(line) for line in lines]

seen_2 = set()
regions_2 = []
perims = []
for r in range(len(grid)):
    for c in range(len(grid[0])):
        if grid[r][c] == -1:
            continue
        else:
            region_start = grid[r][c]
            perim = islandPerimeter(region_start, r, c)
            for cell in seen_2:
                grid[cell[0]][cell[1]] = -1
            perims.append((perim, region_start, r, c))
perims
```
