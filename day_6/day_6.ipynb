```python
# with open("test_input.txt") as f:
with open("input.txt") as f:
    lines = f.readlines()
# lines
```

```python
grid = [list(line.strip()) for line in lines]
grid
```

```python
rows = len(grid)
cols = len(grid[0])
```

```python
def find_guard(grid):
    for i in range(rows):
        for j in range(cols):
            if grid[i][j] in "^v<>":
                return (i, j, grid[i][j])
    return -1


find_guard(grid)
```

```python
directions = {
    "^": (-1, 0),
    "v": (1, 0),
    ">": (0, 1),
    "<": (0, -1),
}
facings = {
    (-1, 0): "^",
    (1, 0): "v",
    (0, 1): ">",
    (0, -1): "<",
}

next_direction = {
    "^": ">",
    ">": "v",
    "v": "<",
    "<": "^",
}
```

```python
guard_x, guard_y, facing = find_guard(grid)
direction = directions[facing]
guard_x, guard_y, direction
```

```python
def print_grid(grid):
    for row in grid:
        print(row)

# print_grid(grid)
```

```python
def print_grid_2(grid):
    fin = []
    for row in grid:
        fin.append("".join(row))

    print("\n".join(fin))


# print_grid_2(grid)
```

```python
while True:
    new_x, new_y = guard_x + direction[0], guard_y + direction[1]
    # print(guard_x, guard_y, direction, facing)
    # print(new_x, new_y)
    if new_x < 0 or new_x >= rows or new_y < 0 or new_y >= cols:
        grid[guard_x][guard_y] = "X"
        break
    while grid[new_x][new_y] == "#":
        facing = next_direction[facing]
        direction = directions[facing]
        new_x, new_y = guard_x + direction[0], guard_y + direction[1]
    # print_grid(grid)
    # print()
    grid[new_x][new_y] = facings[direction]
    grid[guard_x][guard_y] = "X"
    guard_x, guard_y = new_x, new_y

print_grid(grid)
```

```python
def count_x(grid):
    count = 0
    for i in range(rows):
        for j in range(cols):
            if grid[i][j] == "X":
                count += 1
    return count


count_x(grid)
```
