```python
def dijkstra(grid, start, end, rows, cols):
    directions = [(-1, 0), (0, 1), (1, 0), (0, -1)]
    start_x, start_y = start
    start_face = (0, 1)
    start_cost = 0

    # Priority queue and visited dictionary
    pq = []
    heapq.heappush(pq, ((start_cost, start_x, start_y, start_face)))
    visited = {(start_x, start_y, start_face): 0}

    while pq:
        cost, x, y, face = heapq.heappop(pq)

        # Skip if we've already processed a better cost for this state
        if visited.get((x, y, face), float("inf")) < cost:
            continue

        # Move forward
        new_x, new_y = x + face[0], y + face[1]
        if 0 <= new_x < rows and 0 <= new_y < cols and grid[new_x][new_y] != "#":
            new_cost = cost + 1
            if new_cost < visited.get((new_x, new_y, face), float("inf")):
                visited[(new_x, new_y, face)] = new_cost
                heapq.heappush(pq, (new_cost, new_x, new_y, face))

        # Turn left or right
        dir_index = directions.index(face)
        for new_dir in [(dir_index - 1) % 4, (dir_index + 1) % 4]:
            new_cost = cost + 1000
            new_face = directions[new_dir]
            if new_cost < visited.get((x, y, new_face), float("inf")):
                visited[(x, y, new_face)] = new_cost
                heapq.heappush(pq, (new_cost, x, y, new_face))

    return visited
```

```python
weights = dijkstra(grid, start, end, rows, cols)
weights
```
